Ideally, our train and validation datasets should have similar class ratios (class distributions), especially in binary or multiclass classification tasks.
Here’s why:

Representative Evaluation: The validation set is used to estimate how well your model will perform on unseen data. If its class distribution is very different from the training set (or from the real-world data you expect), your validation metrics (accuracy, precision, recall, etc.) may not reflect true performance.
Consistent Learning: If the model is trained on one class ratio and validated on another, it might overfit to the training distribution and underperform on the validation set, or vice versa.

Do we have a class imbalance issue?

A 1:2 ratio is not perfectly balanced, but it’s not severe. Severe imbalance is more like 1:10, 1:50, or even 1:100, where one class is extremely rare. With a 1:2 ratio, most models and loss functions can still learn well, especially if you monitor metrics like precision, recall, and F1-score for both classes. We should keep this distribution, especially when in reflects the real-world.


```python
# Display sammple images from each class

# Get a list of file
benign_files = os.listdir('data/raw/test/0')[:3]
malignant_files = os.listdir('data/raw/test/1')[:3]

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for col in range(3):
    # Benign images
    benign_img = plt.imread(f"data/raw/test/0/{benign_files[col]}")
    axes[0, col].imshow(benign_img, cmap='gray')
    axes[0, col].set_title('Class 0 (Benign)')
    axes[0, col].axis('off')
    
    # Malignant images 
    malignant_img = plt.imread(f"data/raw/test/1/{malignant_files[col]}")
    axes[1, col].imshow(malignant_img, cmap='gray')
    axes[1, col].set_title('Class 1 (Malignant)')
    axes[1, col].axis('off')

plt.tight_layout()
plt.show()
```
